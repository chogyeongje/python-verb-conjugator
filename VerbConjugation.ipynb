```python
from html.parser import HTMLParser
import requests

# extract any pronouns and conjugated verbs from a verbix
# conjugation table
# Both pronouns and verbs are the only things in "span" elements
class ConjugationTableParser(HTMLParser):
    data = set([]) 
    relevant_tag = False
    
    def handle_starttag(self, tag, attrs):
        if (tag == "span"):
            self.relevant_tag = True;

    def handle_endtag(self, tag):
        self.relevant_tag = False;
        
    def handle_data(self, data):
        if self.relevant_tag:
            print('-'*16)
            print(data)
            for word in data.split():
                print(word)
                self.data.add(word)
        
```

```python
def get_conjugations(lang, verb):    
    url = 'http://api.verbix.com/conjugator/html?language={0}&tableurl=http://tools.verbix.com/webverbix/personal/template.htm&verb={1}'    
    r = requests.get(url.format(lang, verb))
    parser = ConjugationTableParser()
    parser.feed(r.text)
    return set(parser.data)
```

```python
language = "spa"
verb = "ser"
words = get_conjugations(language, verb)
```
